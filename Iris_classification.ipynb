```python
data = pd.read_csv('Iris.csv')
```

```python
data
```

```python
data.isnull().sum()
```

```python
y_train=data['Species']
```

```python
data.drop(labels='Species',axis=1,inplace=True)
```

```python
x_train=data.values[0:150]
x_test=data.values[90:]
```

```python
from sklearn.model_selection import train_test_split
x_train_sub,x_test_sub,y_train_sub,y_test_sub=train_test_split(x_train,y_train,random_state=1)
```

```python
from sklearn.ensemble import GradientBoostingClassifier,AdaBoostClassifier
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
learning_rates = [0.05, 0.1, 0.25, 0.5, 0.75, 1]
for learning_rate in learning_rates:
    gb = GradientBoostingClassifier(n_estimators=20, learning_rate = learning_rate, max_features=5, max_depth = 2, random_state = 0)
    gb.fit(x_train_sub, y_train_sub)
    print("Learning rate: ", learning_rate)
    print("Accuracy score (training): {0:.3f}".format(gb.score(x_train_sub, y_train_sub)))
    print("Accuracy score (validation): {0:.3f}".format(gb.score(x_test_sub, y_test_sub)))
    print()
```

```python
learning_rates = [0.05, 0.1, 0.25, 0.5, 0.75, 1]
for learning_rate in learning_rates:
    ad = AdaBoostClassifier()
    ad.fit(x_train_sub, y_train_sub)
    print("Learning rate: ", learning_rate)
    print("Accuracy score (training): {0:.3f}".format(ad.score(x_train_sub, y_train_sub)))
    print("Accuracy score (validation): {0:.3f}".format(ad.score(x_test_sub, y_test_sub)))
    print()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
rfc=RandomForestClassifier(criterion='entropy',random_state=42)
rfc.fit(x_train_sub,y_train_sub)
```

```python
y_pred_train=rfc.predict(x_train_sub)
print(classification_report(y_train_sub,y_pred_train))
```

```python
y_pred_test=rfc.predict(x_test_sub)
print(classification_report(y_test_sub,y_pred_test))
```
